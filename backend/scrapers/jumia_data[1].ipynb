```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Selenium options
options = Options()
# options.add_argument("--headless")  # Remove this to see the browser while scraping
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920x1080")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Initialize WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open Konga smartphones category page
pages=[x  for x in range(1,51)]
smartphones = []
for page in pages:
    URL = f"https://www.konga.com/category/smartphones-7539?page={page}"
    driver.get(URL)
    
    # Wait for elements to load dynamically
    wait = WebDriverWait(driver, 10)
    
    # List to store scraped data
    
    
    # Function to extract product details
    def scrape_page():
        products = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div[class*="_4941f_1HCZm"]')))
    
        for product in products:
            try:
                name = product.find_element(By.TAG_NAME, "h3").text
                link = product.find_element(By.TAG_NAME, "a").get_attribute("href")
                price = product.find_element(By.CSS_SELECTOR, "span[class*='d7c0f_sJAqi']").text
                discount = product.find_element(By.CSS_SELECTOR, "span[class*='_6c244_q2qap']").text if product.find_elements(By.CSS_SELECTOR, "span[class*='_6c244_q2qap']") else "No discount"
                ratings=product.find_element(By.CLASS_NAME, "star").text if product.find_elements(By.CLASS_NAME, "star") else '0')
                
                smartphones.append({
                    "Name": name,
                    "Price": price,
                    "Discount": discount,
                    "Ratings":ratings,
                    "Link": link
                })
    
                print(f"Scraped: {name}")
            except Exception as e:
                print("Error extracting product:", e)
    
    # Scrape all pages
    while True:
        scrape_page()
    
        # Check if "Next" button exists
        try:
            next_button = wait.until(EC.presence_of_element_located((By.XPATH, '//a[contains(@aria-label, "Next page")]')))
            next_button.click()
            time.sleep(3)  # Allow time for new page to load
        except:
            print(f"Page{page} scraped successfully")
            break
    
    # Close browser
driver.quit()

# Print total products scraped
print(f"\nTotal Smartphones Scraped: {len(smartphones)}")
```

```python
import pandas as pd

# Convert list to DataFrame
df = pd.DataFrame(smartphones)

# Save as CSV
df.to_csv("konga_smartphones.csv", index=False, encoding="utf-8")

print("Data saved successfully!")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Selenium options
options = Options()
# options.add_argument("--headless")  # Remove this to see the browser while scraping
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920x1080")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
# Open Konga Phones & Tablets page
url = "https://www.konga.com/category/phones-tablets-5294"
driver.get(url)

# Allow page to load
time.sleep(5)

# Extract product ratings
ratings = []
products = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div[class*="_4941f_1HCZm"]')))

for product in products:
    try:
        rating = product.find_element(By.CSS_SELECTOR, "star").text  # Adjust based on actual class
        ratings.append(rating)
    except:
        ratings.append("No rating")

# Print extracted ratings
for i, rating in enumerate(ratings, 1):
    print(f"Product {i}: {rating}")

# Close browser
driver.quit()
```
